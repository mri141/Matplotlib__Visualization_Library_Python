# Line Graph

```python
x = np.linspace(0,20)
y= x**2
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0,20)
y= x**2
plt.plot(x, y)
plt.xlabel('x-values')
plt.ylabel('x^2-values')
plt.title('line graph of y=Square of x')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10,10)
y= x**2
z= x**3
plt.plot(x,z,color='black')
plt.plot(x, y, '*')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('line plot')
plt.grid(True)
plt.show()
```

```python
plt.plot(np.sin(x),label='sin(x)',color='orange')
plt.plot(np.cos(x),label='cos(x)',color='green')
plt.xlim(0,150)
plt.legend()
plt.title('math functions')
plt.show()
```

```python
plt.plot(np.sin(x),label='sin(x)',color='orange')
plt.plot(np.cos(x),label='cos(x)',color='green')
plt.plot(np.tan(x),label='tan(x)',color='red')
plt.xlim(0,90)
plt.legend()
plt.title('math functions')
plt.show()
```

# Sub plot

```python
fig, axs = plt.subplots(2, 2,gridspec_kw={'hspace': 0.5, 'wspace': 0.5})
x = np.linspace(0, 20, 400)
y = np.sin(x)
z=np.cos(x)
m = (x**3)
n=(x**2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('sin(x)')
axs[0, 1].plot(x, z, 'tab:orange')
axs[0, 1].set_title('cos(x)')
axs[1, 0].plot(x, m, 'tab:green')
axs[1, 0].set_title('x**3')
axs[1, 1].plot(x,n, 'tab:red')
axs[1, 1].set_title('x**2')
plt.show()
```

```python
fig, axs = plt.subplots(2, 5 ,gridspec_kw={'hspace': 0.2, 'wspace': 0.6})
x = np.linspace(0, 20, 400)
y = np.sin(x)
z=np.cos(x)
m = (x**3)
n=(x**2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('sin(x)')
axs[0, 1].plot(x, z, 'orange')
axs[0, 1].set_title('cos(x)')
axs[1, 0].plot(x, m, 'green')
axs[1, 0].set_title('x**3')
axs[1, 1].plot(x,n, 'red')
axs[1, 1].set_title('x**2')
plt.show()
```

# Scatter plots

```python
iris.plot(kind='scatter', x='petal.length', y='petal.width') 
plt.show()
```

```python
iris.plot(kind='line', x='sepal.length', y='sepal.width') 
plt.show()
```

```python
iris.plot(kind='scatter', x='petal.length', y='petal.width',color='green',grid='True',title='Scatter Plot') 
plt.show()
```

```python
colours = {'Setosa':'orange', 'Versicolor':'green', 'Virginica':'blue'}
for i in range(len(iris['sepal.length'])):
  plt.scatter(iris['petal.length'][i],iris['petal.width'][i], color = colours[iris['variety'][i]])
plt.title('Scatter Plot')
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.grid(True)
plt.show()
```

# Bar plots

```python
a= iris['variety'].value_counts()
species_types = a.index
count = a.values
plt.bar(species_types ,count,color = 'red')
plt.xlabel('Type of Species')
plt.ylabel('count')
plt.show()
```

# Boxplot

```python
length_width = iris[['petal.length','petal.width','sepal.length','sepal.width']] #excluding species column
length_width.boxplot()
plt.xlabel('Flower measurements')
plt.ylabel('values')
plt.title("Iris dataset analysis")
plt.show()
```

# Histograms

```python
data_ = np.random.randn(1000)
plt.hist(data_,bins = 50,color='blue') #Binning Procedure
plt.grid(True)
plt.xlabel('points')
plt.title("Histogram")
plt.show()
```

```python
iris.plot.hist(subplots=True, layout=(2,2), figsize=(10, 20), bins=30)
```

# Pie charts

```python
a= iris['variety'].value_counts()
species = a.index
count = a.values
colors= ['lightblue','lightgreen','gold']
explode = (0.0,0.5,0.2) #catagory seperate from other type
plt.pie(count, labels=species,shadow=True,
        colors=colors,explode = explode, autopct='%1.1f%%')
plt.xlabel('species')
plt.axis('equal')
plt.show()
```

# Seaborn(Visualization Library)

```python
import seaborn as sns
```

# Line Graph

```python
sns.set_style('darkgrid')
sns.lineplot(data=iris.drop(['variety'], axis=1))
plt.show()
```

# Scatter Plot

```python
sns.FacetGrid(iris, hue="variety", height=4).map(plt.scatter, "sepal.length", "sepal.width").add_legend()
plt.show()
```

# Bar Plot

```python
a= iris['variety'].value_counts()
species = a.index
count = a.values
sns.barplot(x = species, y=count)
plt.show()
```

```python
sns.FacetGrid(iris, hue="variety", height=5) \
   .map(sns.distplot, "petal.length") \
   .add_legend()
plt.show()
```
